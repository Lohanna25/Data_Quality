```python
import pandas as pd
import numpy as np
#import pandas_profiling

# Cargar tu conjunto de datos
df = pd.read_csv('dataset.csv')

# 1. Análisis Básico de Datos
info_general = df.info()
resumen_estadistico = df.describe(include='all')

# 2. Manejo de Valores Nulos
valores_nulos = df.isnull().sum()

# 3. Duplicados
duplicados = df[df.duplicated()]

# 4. Análisis de Datos Anómalos o Atípicos
anomalias = df.select_dtypes(include=[np.number]).apply(lambda x: x[(x < x.quantile(0.01)) | (x > x.quantile(0.99))])


# Imprimir resultados o realizar acciones adicionales según sea necesario
print("Información general:\n", info_general)
print("\nResumen estadístico:\n", resumen_estadistico)
print("\nValores nulos:\n", valores_nulos)
print("\nDuplicados:\n", duplicados)
print("\nAnomalías:\n", anomalias)

# El informe detallado se genera en un archivo HTML (informe_calidad_datos.html)
```

```python
def validar_unicidad(datos):
    # Iterar sobre todas las columnas
    for columna in datos.columns:
        # Verificar si hay duplicados en la columna actual
        duplicados = datos.duplicated(subset=columna, keep=False)

        if duplicados.any():
            # Imprimir las filas duplicadas
            filas_duplicadas = datos[duplicados]
            print(f"Se encontraron duplicados en la columna '{columna}':")
            print(filas_duplicadas)
        else:
            print(f"No se encontraron duplicados en la columna '{columna}'.")
```

```python
validar_unicidad(df)
```
